### ***Teorema Limite Central***

Um teorema famoso em estatística que diz: "Quanto maior o número de amostras, mais os dados tendem a uma distribuição normal

```python
plt.figure(figsize=(15, 5))

# Primeiro subplot: Distribuição de Idade de Aposentadoria
plt.subplot(121)
plt.title('Distribuição de Idade de Aposentadoria')
plt.xlabel('Idade de Aposentadoria')
plt.ylabel('Frequência')
plt.hist(idade_aposentadoria, bins=30, alpha=0.7, color='blue')

# Segundo subplot: Alturas dos Funcionários
plt.subplot(122)
plt.hist(altura, bins=30, alpha=0.7, color='green')
plt.title('Alturas dos Funcionários')
plt.xlabel('Altura (cm)')

# Exibindo os subplots
plt.tight_layout()
plt.show()
```

##### **Tirando amostras**

```python
# n == Tamanho/len() da amostra
n = 100

# quantidade de amostras
qtd = 10000

# Criei uma função que faz amostras apenas passando alguns parametros
def reamostragem_media(dados, coluna, n, qtd):
    medias = [dados[coluna].sample(n, replace=True).mean() for _ in range(qtd)]
    return medias
```

```python
# Fazendo a reamostragem com a função
medias_altura = reamostragem_media(altura, 'alturas', n, qtd)
medias_idade_aposentadoria = reamostragem_media(idade_aposentadoria, 'idade', n, qtd)
```

### Teorema Limite central na pratica

```python
print('=========Médias=========')
print('Idade Aposentadoria sem tirar Amostras: ', altura['alturas'].mean()) 
print('Idade Aposentadoria tirando Amostras: ', np.array(medias_altura).mean())
print('\nIdade Aposentadoria tirando Amostras: ', np.array(medias_idade_aposentadoria).mean())
print('Idade Aposentadoria tirando Amostras: ', idade_aposentadoria['idade'].mean())
```

Com os graficos podemos ver melhor a diferença e entender melhor o TLC

Vemos o "formato de Sino" 

```python
plt.figure(figsize=(15, 5))

# Primeiro subplot: Distribuição de Idade de Aposentadoria
plt.subplot(121)
plt.title('Distribuição de Idade de Aposentadoria')
plt.xlabel('Idade de Aposentadoria')
plt.ylabel('Frequência')
plt.hist(medias_idade_aposentadoria, bins=30, alpha=0.7, color='blue')

# Segundo subplot: Alturas dos Funcionários
plt.subplot(122)
plt.hist(medias_altura, bins=30, alpha=0.7, color='green')
plt.title('Alturas dos Funcionários')
plt.xlabel('Altura (cm)')

# Exibindo os subplots
plt.tight_layout()
plt.show()
```
