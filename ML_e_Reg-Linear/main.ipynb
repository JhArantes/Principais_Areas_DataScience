# Linha Reg-Linear

Uma linha de regressão linear é uma ferramenta estatística usada para modelar a relação entre uma variável dependente (ou resposta) e uma ou mais variáveis independentes (ou preditoras). A regressão linear simples envolve apenas uma variável preditora, enquanto a regressão linear múltipla envolve várias variáveis preditoras.


#### Linha Reg-linear feita por um humano

```python
# Aparentemente quanto maior a área do primeiro andar, maior o preço da casa.
# E se quisermos traçar uma linha que melhor representa esse comportamento?
plt.scatter(dados['area_primeiro_andar'], dados['preco_de_venda'])
plt.axline(xy1 = (66, 250000),xy2 = (190, 1800000), color = "red" )
plt.title("Relação entre Preço e Area")
plt.xlabel("Area do primeiro andar")
plt.ylabel("Preço de venda")
```

# MACHINE LEARN

## Treino e Teste

O conjunto de **treinamento** é usado para ajustar o modelo, enquanto o conjunto de **teste** é usado para avaliar seu desempenho em prever preços de casas não vistos durante o treinamento, que auxilia na generalização do modelo.

```python
# import train_test_split
from sklearn.model_selection import train_test_split
```

```python
# Definindo y e X
y = dados['preco_de_venda']
X = dados.drop(columns = 'preco_de_venda')
```

## Dividindo os dados em conjuntos de treinamento e teste. Aqui está um resumo do que cada linha faz:


##### Importação:

    A função train_test_split é importada do módulo sklearn.model_selection.

##### Divisão dos Dados:
1. train_test_split(X, y, test_size=0.3, random_state=196):
2. X: Variáveis independentes (também conhecidas como features ou preditoras).
3. y: Variável dependente (também conhecida como target ou resposta).
4. test_size=0.3: Especifica que 30% dos dados serão usados para o conjunto de teste, enquanto os 70% restantes serão usados para o conjunto de treinamento.
5. random_state=196: Define uma semente para o gerador de números aleatórios, garantindo que a divisão dos dados seja reprodutível. Ou seja, ao usar o mesmo random_state, a mesma divisão dos dados será obtida sempre que o código for executado.

##### Retorno da Função:
A função retorna quatro conjuntos de dados:
1. X_train: Variáveis independentes para o conjunto de treinamento.
2. X_test: Variáveis independentes para o conjunto de teste.
3. y_train: Variável dependente para o conjunto de treinamento.
4. y_test: Variável dependente para o conjunto de teste.


### Objetivo
Dividir os dados em conjuntos de treinamento e teste é um passo essencial em muitos fluxos de trabalho de aprendizado de máquina. O conjunto de treinamento é usado para ajustar (treinar) o modelo, enquanto o conjunto de teste é usado para avaliar o desempenho do modelo em dados não vistos, ajudando a validar a capacidade de generalização do modelo.

```python
#Aplicando o split do y e X
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state= 196)
```

```python
#Dados de treino para usar a fórmula

df_train = pd.DataFrame(data= X_train )
df_train['preco_de_venda'] = y_train
```

# Ajustando o modelo 

A função ols (Ordinary Least Squares) é importada do módulo statsmodels.formula.api. Esta função é utilizada para ajustar modelos de regressão linear.

O objetivo deste código é falar para o modelo que ele deve criar a relação entre a var resposta e a var explicativa

```python
# import ols
from statsmodels.formula.api import ols
```

```python
# ajustando o primeiro modelo
modelo_0 = ols('preco_de_venda ~ area_primeiro_andar', data = df_train ).fit()
```
